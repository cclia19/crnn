```python
#concatenate image (SAME carplate)

import cv2
folder = r'C:\Users\Cecil\holmeyoung-crnn-pytorch\crnn-pytorch\a'

images = []
for filename in os.listdir(folder):
    img = cv2.imread(os.path.join(folder,filename))
    if img is not None:
        images.append(img)

names = []
for filename in os.listdir(folder):
    name = os.path.splitext(filename)[0]
    names.append(name)

i=0    
for image in images:
        image_v = cv2.vconcat([image, image])
        os.chdir(r"C:\Users\Cecil\holmeyoung-crnn-pytorch\crnn-pytorch\aa")
        filename = "%s%s_0.jpg" % (names[i], names[i])
        cv2.imwrite(filename, image_v)
        i+=1
```

```python
#shuffle rows in text file

import random
with open(r'C:\Users\Cecil\Desktop\Ori_ImagePath.txt','r') as source:
    data = [ (random.random(), line) for line in source ]
data.sort()
with open(r'C:\Users\Cecil\Desktop\Shuffled_ImagePath.txt','w') as target:
    for _, line in data:
        target.write( line )
```

```python
#concatenate image (MIX carplate)

import cv2
folder = r'C:\Users\Cecil\holmeyoung-crnn-pytorch\crnn-pytorch\a'

shuffleFile = open(r'C:\Users\Cecil\Desktop\Shuffled_ImagePath.txt','r')
lines = shuffleFile.readlines()


images = []
names = []

for line in lines:
    filename = line.rstrip()
    basename = filename.replace(".jpg", "")
    names.append(basename)
    img = cv2.imread(os.path.join(folder,filename))
    if img is not None:
        images.append(img)

i=0   
while i < (len(images)-1):
    image_v = cv2.vconcat([images[i], images[i+1]])
    os.chdir(r"C:\Users\Cecil\holmeyoung-crnn-pytorch\crnn-pytorch\aa")
    filename = "%s%s_0.jpg" % (names[i], names[i+1])
    cv2.imwrite(filename, image_v)
    i+=1
```
